#### import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt

```python
df = yf.download("CCCX")
```

```python
returns = np.log(1 + df['Close', 'CCCX'].pct_change())
```

```python
mu,sigma = returns.mean(), returns.std()
```

```python
sim_rets = np.random.normal(mu,sigma,252)
```

```python
inital = df['Close', 'CCCX'].iloc[-1]
```

```python
sim_prices = inital * (sim_rets + 1).cumprod()
```

```python
plt.plot(sim_prices)
```

```python
for i in range(100):
    sim_rets = np.random.normal(mu,sigma,252)
    sim_prices = inital * (sim_rets + 1).cumprod()
    plt.axhline(inital,c='k')
    plt.plot(sim_prices)
```
